Beaglebone boot proccess:

1. POwer applied. Start program counter at defined location in boot rom.

2. FIRST STAGE BOOTLOADER. CPU boot. Minumum peripheral configuration, finds boot image, loads x-loader.

3. SECOND STAGE BOOTLOADER. X-LOADER. Sets up pin mixing, initializes clocks and memory. Loads U-boot.

4. THIRD STAGE BOOTLOADER. U-BOOT.  Specifies root file system. Uses eEnv.txt configuration. Performs additional initialization. Loads and passes control to linux kernel.

5. Linux Kernel. Unzip the kernel into memory. Sets up peripherals USB, I2C, HDMI. Mounts file system that contains linux applications. 

6. Calls upper space process - init. Moves kernel context to the user context.

## Sysfs

It is a feature of linux kernel that allows kernel code to export information to user via an in- memory filesystem. Has some strict organization. Mostly ascii files, one value per file. Something to do with files and folders of linux

Apparently everything from sysfs is stored in the sys directory. Apparently it tries to detect everything when the thing boots up. And places everything it found there.

<img src="8_sysfs.png">

## contents

### Block

subdirectories for each device that has been discovered. In each directory there are trributes that describe many things including size of the device and the dev_t number. Symbolic link points to the phisical device that the block device maps to. 

There is a directory that exposes and interface to the I/O scheduler. This provides stattistics about device request queue amd some tunable features. Features are used by user or amdmin to tune performance and change I/O scheduler.


### Bus

<img src="8_block.png">

Holds subdirectories for each physical bus type that is support registered in the kernel. Each bus type has two subdirectories: devices and drivers. 

#### Devices 

The _devices_ contain a flat listing of every device discovered on that type of bus in the entire system.

The devices listed are links to the global device tree.

#### Drivers

Contains directories for each device driver that has been registered with the bus type. In each driver directory are files that allow viewing and manipulation of driver parameters.


#### Class

Contains representation of every device class that is registered withe the kernel. Links to device and driver directories. Each class might link to multiple kernel objects. Atributes and parameters that control the object.



#### FIrmware

Intefaces for firmware. Code that is excuted once the system powers on. Specific objects for firmware.

#### Module

Modules that are loaded. Every module represented. Sections directory contains attributes for module. Used in debugging. Refcnt displaysthe reference count or number of users of the module.

#### Power

COntains only two attibutes. Disk controls suspension to disk. State which lets proccess enter low power state, reading it shows what states it supports.. 



## Conclusion:

Sysfs is a virtual file systems which provides interface to the kernel data structure. Provides information about kernel modules, devices. Mounted at /sys. Allows changing of kernel variables.








# Session 9 I2C

Communication bus on PCB

- Speed 100 KHZ
- Master/ slave architecture.

Why does it exist?

- Not much space for more pins on chips.
- Complexity of circut boards. 
- electronic interferance.
- Higher cost for more pins.

Solution:

Simple serial communication bus. Replaced parrallel connection with a lot of pins for non time critical tasks. EEPROM, sensors, clock.

<img src="9_master.png">

It has a two wire interface. SCL is controlled by the master. SDA is bidirectional. Its byte oriented. 

Addressing the devices. Each slave has a 7 bit address. In some slave devices it is hardcoded. Like HIH8120 has 0x27 (0011011). Some of them are configurable. 

## How it works:

READ = 1
WRITE = 0

Master initiates the communication with a start bit. Only one party can comunicate at a time (Like walkie talkie, duplex). Direction bit is set by the master. First byte that is sent holds 7 bit slave address + the read write address. Each recieved byte is acknowledged by either slave or master. Depends on write or read. After last bit the master sets a stop bit.

<img src="9_comunicate.png">

## Request data from slave.

- Master initiates communication with start bit
- Master sends 7 bit slave addres and sets read bit.
- Slave ACK and transmits first byte. Master ACK byte.
- Master does not acknowledge last byte because he doesnt want to read more.

<img src="9_readSlave.png">

## Read specific address in slave.

- Master initiates a write session and transmits the slave address. Sends the address in slave to read.
- Master sends reapeated start with R/W set to read (1).
- Slave sends the data starting form the address the master gave.
- Master ACK the bytes except last one.

<img src="9_readSpecific.png">

## Multiple I/O connected to the same line.

Both input and output are on the same SDA line. The clock determines how often they communicate. When not in use the lines are grounded by pullup resistor.
